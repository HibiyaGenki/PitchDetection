```python
import json
import matplotlib.pyplot as plt

# JSONデータをPythonオブジェクトとして読み込む
with open('/home/ubuntu/slocal/PitchDetection/test_original_predictions.json', 'r') as file:
    data = json.load(file)

# 正規化の計算を行う関数
def normalize(values):
    # 数値のみをフィルタリング
    numeric_values = [val for val in values if isinstance(val, (int, float))]
    if not numeric_values:
        return []  # 数値がない場合は空リストを返す
    min_val = min(numeric_values)
    max_val = max(numeric_values)
    # 正規化計算: 値が一定の場合はすべて0.0にする
    return [(val - min_val) / (max_val - min_val) if max_val != min_val else 0.0 for val in numeric_values]

# 初期化
result = {}

# データを処理
for key, value_list in data.items():  # 各キーに対して処理
    first_values = []
    second_values = []
    for nested_list in value_list:  # ネストされたリストにアクセス
        if isinstance(nested_list, list):  # データがリストかどうか確認
            for pair in nested_list:  # ペアにアクセス
                if isinstance(pair, list) and len(pair) == 2:  # データが2要素のリストであるか確認
                    # 1つ目と2つ目の値をリストに追加
                    first_values.append(pair[0])
                    second_values.append(pair[1])
                else:
                    print(f"Unexpected pair format in {key}: {pair}")
        else:
            print(f"Unexpected nested_list format in {key}: {nested_list}")
    # 各リストを正規化
    if first_values and second_values:  # データが空でない場合のみ処理
        normalized_first = normalize(first_values)
        normalized_second = normalize(second_values)
        # キーごとに結果を保存
        result[key] = {
            "None": normalized_first,  # クラス0
            "Pitch": normalized_second  # クラス1
        }

        # 時間データを生成（インデックスを仮の時間として扱う）
        time = list(range(len(normalized_first)))
        
        # クラス0（None）のグラフをプロット
        plt.figure(figsize=(20, 4))  # 横長・縦短のグラフ
        plt.bar(time, normalized_first, color='blue', alpha=0.7, label="None")
        plt.title(f"Normalized Probabilities for None ({key})")
        plt.xlabel("Time")
        plt.ylabel("Probability")
        plt.legend()
        plt.grid(axis="y")
        plt.tight_layout()
        plt.savefig(f"{key}_None_bar_plot.png")  # Noneクラスのグラフを保存
        plt.close()
        
        # クラス1（Pitch）のグラフをプロット
        plt.figure(figsize=(20, 4))  # 横長・縦短のグラフ
        plt.bar(time, normalized_second, color='orange', alpha=0.7, label="Pitch")
        plt.title(f"Normalized Probabilities for Pitch ({key})")
        plt.xlabel("Time")
        plt.ylabel("Probability")
        plt.legend()
        plt.grid(axis="y")
        plt.tight_layout()
        plt.savefig(f"{key}_Pitch_bar_plot.png")  # Pitchクラスのグラフを保存
        plt.close()

        # クラス0とクラス1を合わせたグラフをプロット
        plt.figure(figsize=(20, 4))  # 横長・縦短のグラフ
        plt.bar([t - 0.2 for t in time], normalized_first, width=0.4, color='blue', alpha=0.7, label="None")
        plt.bar([t + 0.2 for t in time], normalized_second, width=0.4, color='orange', alpha=0.7, label="Pitch")
        plt.title(f"Combined Probabilities for None and Pitch ({key})")
        plt.xlabel("Time")
        plt.ylabel("Probability")
        plt.legend()
        plt.grid(axis="y")
        plt.tight_layout()
        plt.savefig(f"{key}_Combined_bar_plot.png")  # Combinedグラフを保存
        plt.close()

print("クラスごとの棒グラフとCombined棒グラフをプロットし、PNG形式で保存しました。")
```
