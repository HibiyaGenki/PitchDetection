```python
import cv2
import os
import glob
def extract_frames(video_path, out_base_dir):
    cap = cv2.VideoCapture(video_path)
    # output_dir = "frames/" + os.path.basename(video_path)
    output_dir = os.path.join(out_base_dir, "frame", os.path.splitext(os.path.basename(video_path))[0])
    os.makedirs(output_dir, exist_ok=True)

    if not cap.isOpened():
        print("動画を開くことができませんでした")
        exit()


    frame_count = 0
    saved_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        filename = os.path.join(output_dir, f"frame_{saved_count:05d}.jpg")
        if os.path.exists(filename):
            saved_count += 1
            continue
        
        cv2.imwrite(filename, frame)
        saved_count += 1
        # print(f"{filename} を保存")
        frame_count += 1

    cap.release()


# extract_frames("/home/ubuntu/slocal/PitchDetection_torimi/data/Videos/test_HvsW.mp4")
files = glob.glob("/home/ubuntu/slocal/PitchDetection_torimi/data/Videos/*.mp4")
for file in files:
    print(file)
    extract_frames(file, out_base_dir="data")
```
